```python
import csv

movies = []
bookings = []
movies_file = 'movies.csv'
bookings_file = 'bookings.csv'

class Movie:
    def _init_(self, Sr_no, title, duration, genre, time_slots, available_seats):
        self.Sr_no = Sr_no
        self.title = title
        self.duration = duration
        self.genre = genre
        self.time_slots = time_slots
        self.available_seats = available_seats

class Booking:
    def _init_(self, customer_name, movie_title, time_slot, num_tickets):
        self.customer_name = customer_name
        self.movie_title = movie_title
        self.time_slot = time_slot
        self.num_tickets = num_tickets

    def add_movie(Sr_no, title, duration, genre, time_slots, available_seats):
        movie = Movie(Sr_no, title, duration, genre, time_slots, available_seats)
        movies.append(movie)

    def save_movies_to_csv():
        with open(movies_file, 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["Sr_no", "Title", "Duration", "Genre", "Time", "Available Seats"])
            for movie in movies:
                writer.writerow([movie.Sr_no, movie.title, movie.duration, movie.genre, movie.time_slots, movie.available_seats])

    def load_movies_from_csv():
        try:
            with open(movies_file, 'r') as file:
                reader = csv.reader(file)
                next(reader)  # Skip the header row
                for row in reader:
                    Sr_no, title, duration, genre, time_slots, available_seats = row
                    add_movie(int(Sr_no), title, duration, genre, time_slots, int(available_seats))
        except FileNotFoundError:
            pass

    def display_movies():
        for movie in movies:
            print("Sr_no:", movie.Sr_no)
            print("Title:", movie.title)
            print("Duration:", movie.duration)
            print("Genre:", movie.genre)
            print("Time:", movie.time_slots)
            print("Available Seats:", movie.available_seats)
            print("---------------")
    add_movie(1, "Inception", "2h 28m", "Sci-Fi", "10am, 1pm, 4pm, 7pm, 10:00pm", 100)
    add_movie(2, "The Dark Knight", "2h 32m", "Action", "10:00am, 1:00pm, 4:00pm, 7:00pm, 10:00pm", 50)
    add_movie(3, "Evil Dead Rise", "2h ", "Action", "10:00am, 1:00pm, 4:00pm, 7:00pm, 10:00pm", 100)

    def make_booking(customer_name, movie_title, time_slot, num_tickets):
        for movie in movies:
            if movie.title == movie_title:
                if movie.available_seats >= num_tickets:
                    booking = Booking(customer_name, movie_title, time_slot, num_tickets)
                    bookings.append(booking)
                    movie.available_seats -= num_tickets
                    save_bookings_to_csv()
                    save_movies_to_csv()
                    return
                else:
                    print("Insufficient seats!")
                    return
        print("Movie not found!")

    def save_bookings_to_csv():
        with open(bookings_file, 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["Customer Name", "Movie Title", "Time_slot", "Number of Tickets"])
            for booking in bookings:
                writer.writerow([booking.customer_name, booking.movie_title, booking.time_slot, booking.num_tickets])

    def load_bookings_from_csv():
        try:
            with open(bookings_file, 'r') as file:
                reader = csv.reader(file)
                next(reader)  # Skip the header row
                for row in reader:
                    customer_name, movie_title, time_slot, num_tickets = row
                    booking = Booking(customer_name, movie_title, time_slot, int(num_tickets))
                    bookings.append(booking)
        except FileNotFoundError:
            pass

    def display_bookings():
        for booking in bookings:
            print("Customer Name:", booking.customer_name)
            print("Movie Title:", booking.movie_title)
            print("Time_slot:", booking.time_slot)
            print("Number of Tickets:", booking.num_tickets)
            print("---------------")

display_movies()
def main():
    load_movies_from_csv()
    load_bookings_from_csv()

    seats1 = list(range(1, 101))
    seats2 = list(range(1, 51))
    seats3 = list(range(1, 101))

    

    while True:
        print("Welcome to Book Your Show")
        print("Please enter the required details in order to have a hassle-free movie time!!")
        ch = int(input("Enter Choice of Movie: "))
        
        if ch == 1:
            print(movies[0].title)
            customer_name = input("Enter Your Name: ")
            movie_title = input("Enter the movie title: ")
            time_slot = input("Enter the time: ")
            num_tickets = int(input("Enter the number of tickets: "))

            print("SEATS AVAILABLE:", seats1)
            for i in range(num_tickets):
                seat_no = int(input("Enter seat no: "))
                seats1.remove(seat_no)

            make_booking(customer_name, movie_title, time_slot, num_tickets)
            print(" ")
            print(customer_name)
            print(movie_title)
            print(time_slot)
            print(num_tickets)
            print(" ")
            print("Booking Confirmed!")
            print("Thank You For Booking")
            print("Enjoy Your Movie!")
            print("--------------------")
            print(" ")

        elif ch == 2:
            print(movies[1].title)
            customer_name = input("Enter Your Name: ")
            movie_title = input("Enter the movie title: ")
            time_slot = input("Enter the time: ")
            num_tickets = int(input("Enter the number of tickets: "))

            print("SEATS AVAILABLE:", seats2)
            for i in range(num_tickets):
                seat_no = int(input("Enter seat no: "))
                seats2.remove(seat_no)

            make_booking(customer_name, movie_title, time_slot, num_tickets)
            print(" ")
            print(customer_name)
            print(movie_title)
            print(time_slot)
            print(num_tickets)
            print(" ")
            print("Booking Confirmed!")
            print("Thank You For Booking")
            print("Enjoy Your Movie!")
            print("--------------------")
            print(" ")

        elif ch == 3:
            print(movies[2].title)
            customer_name = input("Enter Your Name: ")
            movie_title = input("Enter the movie title: ")
            time_slot = input("Enter the time: ")
            num_tickets = int(input("Enter the number of tickets: "))

            print("SEATS AVAILABLE:", seats3)
            for i in range(num_tickets):
                seat_no = int(input("Enter seat no: "))
                seats3.remove(seat_no)

            make_booking(customer_name, movie_title, time_slot, num_tickets)
            print(" ")
            print(customer_name)
            print(movie_title)
            print(time_slot)
            print(num_tickets)
            print(" ")
            print("Booking Confirmed!")
            print("Thank You For Booking")
            print("Enjoy Your Movie!")
            print("--------------------")
            print(" ")

if _name_ == '_main_':
    main()
```
